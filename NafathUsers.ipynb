```python
import matplotlib.pyplot as plt
```

```python
# Data
total_users = 21356291
male_users = 15759559
female_users = 5596796
```

```python
# Labels
labels = ['Total Users', 'Male Users', 'Female Users']
```

```python
# Colors
colors = ['blue', 'skyblue', 'pink']
```

```python
# Explode
explode = [0.1, 0, 0]  # Slightly separate the total slice
```

```python
# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie([total_users, male_users, female_users], labels=labels, colors=colors, autopct="%1.1f%%", startangle=140, explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
```
